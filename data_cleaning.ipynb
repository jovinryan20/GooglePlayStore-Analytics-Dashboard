```python
import pandas as pd
```

```python
# Importing dataset
data = pd.read_csv("data/Play Store Data.csv")
```

```python
# Cleaning 'Rating' column
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')
data = data.dropna(subset=['Rating'])
```

```python
# Cleaning 'Installs' and 'Reviews' column
data['Installs'] = data['Installs'].astype(str).str.replace('[+,]','', regex = True)
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')
data['Reviews'] = pd.to_numeric(data['Reviews'], errors='coerce')
data = data.dropna(subset=['Installs', 'Reviews'])
```

```python
# Cleaning 'Price' column
data['Price'] = data['Price'].astype(str).str.replace('[$]','', regex = True)
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
```

```python
# Cleaning 'Android ver' column 
data["Android Ver"] = (
    data["Android Ver"].astype(str).str.extract(r"(\d+\.\d+)")[0].astype(float))
data["Android Ver"] = pd.to_numeric(data["Android Ver"], errors="coerce")
```

```python
# Cleaning the size

def convert_size(size):
    size = str(size).strip()
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size or 'K' in size:
        return float(size.replace('k', '').replace('K', '')) / 1024
    else:
        return None
    
data['Size'] = data['Size'].apply(convert_size)
data.rename(columns={'Size': 'Size_MB'}, inplace=True)
data = data.dropna(subset=['Size_MB'])
data["Size_MB"] = pd.to_numeric(data["Size_MB"], errors="coerce")
```

```python
# #Adding 'Revenue' column
data['Revenue'] = data['Price'] * data['Installs']
```

```python
# Cleaning 'Last Updated' colomn

data['Last Updated'] = pd.to_datetime(data['Last Updated'], errors='coerce')
data = data.dropna(subset=['Last Updated'])
```

```python
# Saving cleaned data to csv

data.to_csv("cleaned_playstore1.csv", index=False)
print("Cleaned data saved as cleaned_playstore.csv")
```
